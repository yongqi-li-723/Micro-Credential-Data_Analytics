```
# set up environment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot

%matplotlib inline

# import data
df = pd.read_csv("http://avijitroy.com/microCred6/covid.csv")

df
```

```
# group by state
state_group = df.groupby(['state']).count().sort_values(by=['cases'], ascending=True)
state_group
```

```
# 2. check NY state data
ny_state = df[df['state'] == 'New York']
ny_state
```

```
# 3. sort NY state data by county
ny_state.sort_values(by=['county'], ascending=True)
```

```
# 4. NYC cases
nyc = df[(df['county'] == 'New York City')]
nyc
```

```
nyc['date'].is_unique
```

```
# 5
nyc['daily cases'] = nyc.cases.diff()
nyc['daily deaths'] = nyc.deaths.diff()
nyc
```

```
nyc
```

```
nyc_30days = nyc.sort_values(by=['date'],ascending=True)
nyc_30days
```

```
# the recent record in the past 30 days, which are at the tail of the list
nyc_30days.tail(31)
```
